```python
ec = EllipticCurve(-2, 4)
p_q = ec.add((0, 2), (3, -5), faction=True)
p_p = ec.add((0, 2), (0, 2))
q_q = ec.add((3, -5), (3, -5))
r_s = ec.add((0, 2), ec.INF)
print("P + Q = ", p_q)
print("P + P = ", p_p)
print("Q + Q = ", q_q)
print("R + S = ", r_s)
```

```python
ec2 = EllipticCurve(0, 17)
p_q2 = ec2.add((-1, 4), (2, 5), faction=True)
print("P + Q = ", p_q2)

p__q2 = ec2.add((-1, 4), (2, -5), faction=True)
print("P + -Q = ", p__q2)

p_p2 = ec2.add((-1, 4), (-1, 4), faction=True)
print("P + P = ", p_p2)

q_q2 = ec2.add((2, 5), (2, 5), faction=True)
print("Q + Q = ", q_q2)

ec2.plot_elliptic_curve()
```
